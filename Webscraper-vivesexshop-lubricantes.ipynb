```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}

productlinks = []

for x in range(1,14):
    r = requests.get(f'https://www.vivesexshop.com/es/aceites-y-lubricantes?pagina={x}')
    soup = BeautifulSoup(r.content, 'lxml', from_encoding="utf-8")
    productlist = soup.find_all('div', class_='col-6 col-sm-4 contenedor-producto')
    for item in productlist:
        for link in item.find_all('a', href=True):
            productlinks.append(link['href'])
```

```python
toy_list = []
for link in productlinks:
    r = requests.get(link, headers=headers)
    soup = BeautifulSoup(r.content, 'lxml')
    try:
        name = soup.find('h1').text.strip()
    except:
        name = 'No name'
     
    try:    
        brand = soup.find('div',class_='col-12 col-md-6 col-lg-5 paddingT20-xs nombre-producto flota-derecha').find('a').get('title')
        
    except:
        brand = 'No brand'
        
    try:    
        price = soup.find('span', class_='precio_web').text.strip()
    except:
        price = "No price"
    
    try:
        oldprice = soup.find('span', class_='precio-tachado').text.strip()
    except:
        oldprice='No old price'
        
    try:    
        ref = soup.find('span', id='referencia').text.strip()
    except:
        ref='No ref'
    
    try:
        rating = soup.find('a', class_='ancla-opiniones ancla-movimiento').text.strip()
    except:
        rating = 'no rating'
        
    url = link
        
    toys = {
        
        'name': name,
        'brand': brand,
        'reference': ref,
        'rating': rating,
        'actual price': price,
        'old price': oldprice,
        'url': url
        }

    toy_list.append(toys)
#     print(brand)
    print('Saved: ', toys['name'])
    
df = pd.DataFrame(toy_list)
df.to_csv('vivesexshop_lubricantes.csv', index=False)
print('All saved!')
```
